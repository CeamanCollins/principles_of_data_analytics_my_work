```python
import sklearn.datasets
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
diabetes = sklearn.datasets.load_diabetes()

diabetes['feature_names']
```

```python
bmi = diabetes['data'][:, 2]
bp = diabetes['data'][:, 3]
sex = diabetes['data'][:, 1]
age = (diabetes['data'][:, 0] + 1) * 30

men = diabetes['data'][diabetes['data'][:, 1] > 0]
men_bmi = men[:, 2]
men_bp = men[:, 3]
women = diabetes['data'][diabetes['data'][:, 1] < 0]
women_bmi = women[:, 2]
women_bp = women[:, 3]
```

```python
age
```

```python
fig, ax = plt.subplots()
plt.scatter(bmi,bp,s=age,c=sex,cmap='tab10')

p1 = np.poly1d(np.polyfit(bmi,bp,1))
xp1 = np.linspace(-0.1,0.175,2)
p2 = np.poly1d(np.polyfit(men_bmi,men_bp,1))
p3 = np.poly1d(np.polyfit(women_bmi,women_bp,1))

plt.plot(xp1,p1(xp1),color='tab:blue')
plt.plot(xp1,p2(xp1),color='tab:orange')
plt.plot(xp1,p3(xp1),color='tab:green')
```
